# ETL - Projeto Airbnb San Francisco, Califórnia, EUA

## Airbnb – San Francisco, Califórnia, EUA

## 1. Introdução

O Airbnb é uma plataforma global que conecta anfitriões e viajantes, oferecendo opções de hospedagem variadas em diferentes cidades do mundo. No contexto da ciência de dados, a empresa se destaca por incentivar a transparência ao disponibilizar dados públicos por meio do portal Inside Airbnb.

Neste notebook, realizamos uma Análise Exploratória de Dados (EDA) utilizando o conjunto de dados de locações da cidade de San Francisco, Califórnia, EUA — cidade de origem do Airbnb e um dos mercados mais relevantes da plataforma. O objetivo é compreender padrões de preços, disponibilidade, tipos de imóveis e características dos anfitriões, extraindo insights que auxiliem tomadas de decisão baseadas em dados.


## 2. Objetivo do tratamento

Os principais objetivos:

- Compreender a estrutura, conteúdo e qualidade do dataset

- Tratar os dados brutos, corrigindo inconsistências, valores ausentes e tipos de dados

- Construir um dataset relevante e consistente, que sirva de base para as etapas posteriores de análise

- Obter os dados diretamente de uma fonte confiável, o portal Inside Airbnb

- Garantir que os dados representem fielmente o fenômeno estudado, sem distorções analíticas

- Preservar a integridade do dataset bruto, assegurando rastreabilidade e reprodutibilidade da análise
