# Projeto 01 - Análise de Dados - AirBnB - San Francisco, Estados Unidos 

Airbnb é uma empresa global de tecnologia e hospitalidade que revolucionou a forma como as pessoas viajam e se hospedam. Fundada em 2007 em São Francisco, a plataforma surgiu quando dois anfitriões receberam três hóspedes em sua casa e, desde então, cresceu exponencialmente, reunindo mais de 5 milhões de anfitriões em mais de 220 países e regiões, com mais de 8 milhões de anúncios ativos — incluindo casas, apartamentos e experiências únicas oferecidas por pessoas locais.

O modelo de negócios da Airbnb é construído sobre uma plataforma comunitária que conecta viajantes a hospedagens alternativas, permitindo que indivíduos aluguem espaços que vão de quartos simples até propriedades completas, com foco em experiências autênticas e imersivas. A empresa cria valor cobrando uma taxa sobre cada reserva realizada via seu site ou aplicativo, e promove um ecossistema onde anfitriões podem gerar renda extra e viajantes podem explorar destinos de forma mais personalizada e flexível.

Além das hospedagens, a Airbnb tem ampliado seus serviços com iniciativas como Airbnb Experiences, que oferece atividades e serviços locais — como passeios, aulas ou vivências culturais — e, mais recentemente, expandiu seu portfólio para incluir serviços adicionais dentro da plataforma, reforçando sua visão de ser mais do que um simples serviço de acomodações e passando a atuar como uma companhia de viagens e estilo de vida integrada.

Com relação à ciência de dados, uma das iniciativas relevantes do Airbnb é a disponibilização pública de dados de locações de algumas das principais cidades do mundo. Por meio do portal Inside Airbnb, a empresa promove transparência e incentiva análises independentes sobre o impacto da plataforma no mercado imobiliário e no turismo. Esses dados permitem estudos sobre preços, disponibilidade, perfil dos anfitriões e padrões de ocupação. No presente trabalho, serão analisados os dados de locações da cidade de San Francisco, na Califórnia, Estados Unidos, local de origem da empresa e um dos mercados mais relevantes da plataforma.

Pontos de destaque da cidade de San Francisco (Califórnia, EUA)

- **Berço da inovação tecnológica:** San Francisco está inserida no ecossistema do Vale do Silício, sendo um dos principais polos mundiais de tecnologia, startups e inovação, o que influencia diretamente o perfil de viajantes e a demanda por hospedagens de curta duração.

- **Alto custo de vida e mercado imobiliário competitivo:** A cidade possui um dos mercados imobiliários mais caros dos Estados Unidos, tornando o aluguel de curta temporada uma alternativa atrativa tanto para moradores quanto para investidores — fator essencial para análises econômicas e de precificação no Airbnb.

- **Forte apelo turístico:** San Francisco abriga atrações icônicas como a Golden Gate Bridge, Alcatraz, Fisherman’s Wharf e bairros culturais como Chinatown e Mission District, o que gera alta demanda turística ao longo do ano.

- **Diversidade cultural e urbana:** A cidade é reconhecida por sua diversidade cultural, social e arquitetônica, refletindo-se em uma ampla variedade de tipos de hospedagem, perfis de anfitriões e experiências oferecidas na plataforma.

- **Regulamentação rigorosa de aluguéis de curta duração:** San Francisco possui leis específicas para plataformas como o Airbnb, tornando seus dados especialmente relevantes para análises sobre impacto regulatório, compliance e sustentabilidade do modelo de negócios.
